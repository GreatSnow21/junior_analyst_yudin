```python
import pandas as pd
from decimal import Decimal
import matplotlib.pyplot as plt
from datetime import datetime

# Загрузка Excel-файла
data = pd.read_excel("/content/drive/MyDrive/Junior Analyst_Юдин/data.xlsx")

# Выводим таблицу с данными
data
```

```python
# Преобразование строк с датами в формат datetime
data['receiving_date'] = pd.to_datetime(data['receiving_date'], errors='coerce')
```

```python
# Группировка по месяцам
data['month_year'] = 0
count_num = 0
dates_month = {'Май 2021': '2021-05', 'Июнь 2021': '2021-06', 'Июль 2021': '2021-07', 'Август 2021': '2021-08', 'Сентябрь 2021': '2021-09', 'Октябрь 2021': '2021-10'}

for deal_date in data['month_year']:
    name_ind = data.loc[count_num, 'status']
    if name_ind in dates_month:
        imp_month = dates_month[name_ind]
    data.loc[count_num, 'month_year'] = imp_month
    count_num += 1

data['month_year']
```

**1. Общая выручка за июль 2021 по сделкам без просрочки**

```python
july_2021_revenue = data[(data['month_year'] == '2021-07') &
                         (data['status'] != 'ПРОСРОЧЕНО')]['sum'].sum()
print(f"Общая выручка за июль 2021 по сделкам без просрочки: {july_2021_revenue}")
```

**2. Изменение выручки компании за рассматриваемый период**

```python
# Построение графика изменения выручки

monthly_revenue = data.groupby('month_year')['sum'].sum()

plt.figure(figsize=(10, 5))
monthly_revenue.plot(kind='line', marker='o')
plt.title("Изменение выручки компании по месяцам")
plt.xlabel("Месяц")
plt.ylabel("Выручка")
plt.grid(True)
plt.show()
```

**3. Менеджер с наибольшей выручкой в сентябре 2021**

```python
september_2021_data = data[data['month_year'] == '2021-10']
top_manager_september = september_2021_data.groupby('sale')['sum'].sum().idxmax()
print(f"Менеджер с наибольшей выручкой в сентябре 2021: {top_manager_september}")
```

**4. Преобладающий тип сделок в октябре 2021**

```python
october_2021_data = data[data['month_year'] == '2021-10']
deal_type_counts = october_2021_data['new/current'].value_counts()
dominant_deal_type = deal_type_counts.idxmax()
print(f"Преобладающий тип сделок в октябре 2021: {dominant_deal_type}")
```

**5. Оригиналы договора по майским сделкам, полученные в июне 2021**

```python
may_deals = data[data['month_year'] == '2021-06']

june_originals_count = may_deals[may_deals['receiving_date'].dt.month == 5]['document'].count()
print(f"Оригиналы договора по майским сделкам, полученные в июне 2021: {june_originals_count}")
```

**6. Вычисление остатка бонусов на 01.07.2021 для каждого менеджера**

```python
# Условие для фильтрации сделок до 01.07.2021
june_data = data[(data['receiving_date'] <= '2021-06-30')]

# Функция для расчета бонуса по сделке
def calculate_bonus(row):
    if row['new/current'] == 'новая' and row['status'] == 'ОПЛАЧЕНО' and row['document'] == 'оригинал':
        return row['sum'] * 0.07  # 7% за новые сделки
    elif row['new/current'] == 'текущая' and row['document'] == 'оригинал' and row['status'] != 'ПРОСРОЧЕНО':
        return row['sum'] * 0.05 if row['sum'] > 10000 else row['sum'] * 0.03  # 5% или 3% за текущие сделки
    return 0  # Нет бонуса, если условия не выполнены

# Применение функции для подсчета бонусов
june_data['bonus'] = june_data.apply(calculate_bonus, axis=1)

# Группировка бонусов по менеджерам
remaining_bonus = june_data.groupby('sale')['bonus'].sum()
```
